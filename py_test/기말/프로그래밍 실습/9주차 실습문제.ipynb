#### 2번 문제
###### 사용자가 1부터 9까지의 숫자를 무작위로 입력한 후, 1부터 9까지의 숫자가  몇개씩 입력되었는지를 확인할 수 있는 프로그램을 작성하시오.

```python
print("1~9 사이 숫자 입력하시오.")
numbers=input("(입력) : ").split()

numAll=[]
numCnt={}

for i in range(len(numbers)) :
    numAll.append(int(numbers[i]))

for i in range(1,10):
    cnt=0
    for j in range(len(numAll)):
        if i==numAll[j] :
            cnt+=1
    numCnt[i]=cnt

print("(결과) : ",end="")
for k,v in numCnt.items() :
    print(k,":",v, end=", ")
```

#### 3번 문제
###### 문장 속에 포함되어 있는 단어의 빈도수를 구하는 프로그램을 작성하시오.
###### 또 빈도수를 기준으로 내림차순으로 정렬하여 출력해보시오.

```python
contents = """
    My fellow Koreans‚ I am grateful to you all. I bow my head in deep appreciation for the
    choice of you‚ the people. Today‚ serving as President in the 19th presidential term of the Republic
    of Korea‚ I take the first step toward a new Korea. My shoulders are now burdened with heavy
    mandates entrusted to me by the people‚ and my heart is burning with enthusiasm to create
    the kind of country that we have never been able to live in before. My head is now filled with
    blueprints for ushering in a new world characterized by unity and coexistence
"""

word_cnt={}
words=contents.split()

for i in words :
    if i in word_cnt :
        word_cnt[i]+=1
    else :
        word_cnt[i]=1

lst=list(word_cnt.items())
items=[]
for (x,y) in lst :
    items.append([y,x])
    
items.sort(reverse=True)

for i in range(len(items)) :
    print(items[i][1],":",items[i][0])
```

#### 4번 문제
###### (집합 변수) 다음 프로그램의 결과를 예측해보시오.
###### 집합은 순서가 없는 원소의 모임이므로 인덱스와 슬라이싱을 사용할 수 없으며, 중복되는 데이타는 한번으로만 간주한다.

```python
# <기본적인 집합 만들기>
set_a = {1, 2, 3, 4}

# <튜플로부터 집합 만들기>
a = ( 1, 2, 2, 3, 3, 4)
b = set(a)
print(b)
# {1, 2, 3, 4}

# <리스트로부터 집합 만들기>
a = [ 1, 2, 2, 3, 3, 4]
b = set(a)
print(b)
# {1, 2, 3, 4}

# <문자열로부터 집합 만들기>
s = 'hong gil dong'
b = set(s)
print(b)
# {'l', 'd', 'n', 'g', 'i', ' ', 'h', 'o'}

# <집합에서 사용할 수 있는 메소드>
s = {100, 200, 300, 300, 400 }
print(s)
# {200, 100, 400, 300}

s.add(500)
print(s)
# {100, 200, 300, 400, 500}

s.discard(100)
print(s)
# {200, 300, 400, 500}

s.clear()
print(s)
# set()

s1 = { 1, 2, 3}
s2 = {3, 4, 5, 6}
print(s1 | s2)
# {1, 2, 3, 4, 5, 6}
print(s1.union(s2))
# {1, 2, 3, 4, 5, 6}
print(s1 & s2)
# {3}
print(s1.intersection(s2))
# {3}
print(s1 - s2)
# {1, 2}
print(s1.difference(s2))
# {1, 2}
print(s1.issubset(s2))
# False
print(s2.issubset(s1))
# False
```
